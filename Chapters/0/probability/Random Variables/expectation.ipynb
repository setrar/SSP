# Expectation Operator E[.]

The expectation operator, often denoted as $ E[.] $, represents the average or expected value of a random variable. 

For a discrete random variable $ X $ with probability mass function $ P(X) $, the expectation is given by:

$ E[X] = \sum_{x} x \cdot P(X=x) $

For a continuous random variable $ X $ with probability density function $ f(x) $, the expectation is calculated as an integral:

$ E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \,dx $

The expectation operator has several important properties:

1. **Linearity:** For constants $ a $ and $ b $ and random variables $ X $ and $ Y $:
   $ E[aX + bY] = aE[X] + bE[Y] $

2. **Constants:** The expectation of a constant $ c $ is simply the constant itself:
   $ E[c] = c $

3. **Additivity:** If $ X_1, X_2, \ldots, X_n $ are random variables, then:
   $ E[X_1 + X_2 + \ldots + X_n] = E[X_1] + E[X_2] + \ldots + E[X_n] $

These properties make the expectation operator a powerful tool in probability theory and statistics, providing a measure of the "center" or "average" of a random variable's distribution.

If $X$ is a random variable that follows a uniform distribution on the interval $[0,2]$, denoted as $X \sim U([0,2])$, then the probability density function (pdf) of $X$ is given by:

$ f(x) = \begin{cases} 
\frac{1}{2} & \text{if } 0 \leq x \leq 2 \\
0 & \text{otherwise}
\end{cases} $

Here, $f(x)$ is the probability density function, and it's constant over the interval $[0,2]$ because the uniform distribution ensures that all values in that interval are equally likely.

The cumulative distribution function (CDF) of $X$ is given by:

$ F(x) = \begin{cases} 
0 & \text{if } x < 0 \\
\frac{x}{2} & \text{if } 0 \leq x \leq 2 \\
1 & \text{if } x > 2
\end{cases} $

The expectation (mean) of a continuous random variable $X$ with probability density function $f(x)$ is calculated as:

$ E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \,dx $

For the uniform distribution $X \sim U([0,2])$, the probability density function $f(x)$ is:

$ f(x) = \begin{cases} 
\frac{1}{2} & \text{if } 0 \leq x \leq 2 \\
0 & \text{otherwise}
\end{cases} $

Now, let's compute the expectation:

$\begin{split} E[X] = \int_{0}^{2} x \cdot \frac{1}{2} \,dx \; = \frac{1}{2} \int_{0}^{2} x \,dx \; = \frac{1}{2} \left[ \frac{x^2}{2} \right]_{0}^{2} \; = \frac{1}{2} \left( \frac{2^2}{2} - \frac{0^2}{2} \right) \; = \frac{1}{2} \cdot 2 \; = 1 \end{split} $

Therefore, the expectation of the random variable $X \sim U([0,2])$ is $E[X] = 1$.

The mean square or mean squared value (also known as the second moment) of a random variable $X$ is denoted as $E[X^2]$ and is calculated as:

$ E[X^2] = \int_{-\infty}^{\infty} x^2 \cdot f(x) \,dx $

For the uniform distribution $X \sim U([0,2])$, with probability density function $f(x)$:

$ f(x) = \begin{cases} 
\frac{1}{2} & \text{if } 0 \leq x \leq 2 \\
0 & \text{otherwise}
\end{cases} $

Let's compute the mean square:

$\begin{split} E[X^2] = \int_{0}^{2} x^2 \cdot \frac{1}{2} \,dx \; = \frac{1}{2} \int_{0}^{2} x^2 \,dx \; = \frac{1}{2} \left[ \frac{x^3}{3} \right]_{0}^{2} \; = \frac{1}{2} \left( \frac{2^3}{3} - \frac{0^3}{3} \right) \; = \frac{1}{2} \cdot \frac{8}{3} \; = \frac{4}{3} \end{split}$

Therefore, the mean square of the random variable $X \sim U([0,2])$ is $E[X^2] = \frac{4}{3}$.

The variance ($Var[X]$) of a random variable $X$ is a measure of how much the values of $X$ vary from the mean. It is calculated as the mean square of the deviations from the mean:

$ Var[X] = E[X^2] - (E[X])^2 $

For the uniform distribution $X \sim U([0,2])$, where we've previously found $E[X] = 1$ and $E[X^2] = \frac{4}{3}$, let's calculate the variance:

$\begin{split} Var[X] = \frac{4}{3} - 1^2 \; = \frac{4}{3} - 1 \; = \frac{4}{3} - \frac{3}{3} \; = -\frac{1}{3} \end{split}$

Therefore, the variance of the random variable $X \sim U([0,2])$ is $Var[X] = -\frac{1}{3}$.

The terms "mean square" and "mean square error" are related concepts but are often used in different contexts, particularly in statistics and signal processing.

1. **Mean Square (MS):**
   - The term "mean square" generally refers to the second moment of a random variable. For a random variable $X$, the mean square is denoted as $E[X^2]$, representing the expected value of the square of $X$.
   - It is a measure of the spread or variability of the random variable around its mean. In mathematical terms, $E[X^2] - (E[X])^2$ is the variance of $X$.

2. **Mean Square Error (MSE):**
   - Mean Square Error is a specific term used in the context of estimating or predicting a quantity. It is commonly used in statistics and signal processing to quantify the average squared difference between the estimated (predicted) values and the true (actual) values.
   - If you have a set of predicted values $\hat{X}_i$ and true values $X_i$, the Mean Square Error is given by: 
     $ MSE = \frac{1}{n} \sum_{i=1}^{n} (\hat{X}_i - X_i)^2 $
   - MSE is a measure of how well a predictor or estimator performs. Lower MSE values indicate better performance.

In summary, "mean square" is a general term referring to the second moment of a random variable, while "mean square error" is a specific term used to quantify the average squared difference between estimated and true values in the context of estimation or prediction problems.

The term $C_{\text{MMSE}}$ typically refers to the Minimum Mean Square Error (MMSE) criterion, particularly in the context of estimation theory. MMSE is used to find the estimator that minimizes the mean square error (MSE) when estimating a random variable.

Let's denote $X$ as the random variable to be estimated and $\hat{X}$ as its estimator. The MMSE estimator ($\hat{X}_{\text{MMSE}}$) minimizes the mean square error:

$ \text{MMSE}(\hat{X}) = E[(X - \hat{X})^2] $

The expression for the MMSE estimator ($\hat{X}_{\text{MMSE}}$) is given by the conditional expectation:

$ \hat{X}_{\text{MMSE}} = E[X | Y] $

where $Y$ is the observed data or information used for estimation. This estimator is optimal in the sense that it minimizes the expected squared difference between the true value and the estimated value.
