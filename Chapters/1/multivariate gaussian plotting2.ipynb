```julia
using Plots
using Distributions
```

```julia
# Function to create a meshgrid
function meshgrid(x, y)
   X = [x for _ in y, x in x]
   Y = [y for y in y, _ in x]
   X, Y
end
```

```julia
# Set the mean and covariance matrix for the multivariate Gaussian distribution
μ = [0.0, 0.0]  # Mean vector
Σ = [1.0 0.5; 0.5 1.0]  # Covariance matrix

# Create a multivariate Gaussian distribution
mvn = MvNormal(μ, Σ)
```

```julia
# Generate a grid of points for x and y
x = range(-3, stop=3, length=100)
y = range(-3, stop=3, length=100)

# Create a meshgrid from x and y
X, Y = meshgrid(x, y);
```

```julia
# Evaluate the multivariate Gaussian PDF at each point in the meshgrid
Z = [pdf(mvn, [X[i, j], Y[i, j]]) for i in 1:size(X, 1), j in 1:size(X, 2)];
```

```julia
# Plot the surface
surface(x, y, Z, xlabel="X", ylabel="Y", zlabel="PDF", title="Multivariate Gaussian Distribution")
```

```julia
# Plot the contour with a colormap
contour(x, y, Z, xlabel="X", ylabel="Y"
    , title="Contour of Multivariate Gaussian Distribution"
    , color=:viridis
    , c=:auto
)
```
