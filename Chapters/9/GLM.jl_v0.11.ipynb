# GLM Minimal Examples

## Ordinary Least Squares Regression:

```julia
using DataFrames, GLM, Plots
```

```julia
data = DataFrame(X=[1,2,3], Y=[2,4,7])
```

```julia
ols = lm(@formula(Y ~ X), data)
```

```julia
round.(stderror(ols), digits=5)
```

```julia
round.(predict(ols), digits=5)
```

```julia
round(r2(ols); digits=5)
```

```julia
round(loglikelihood(ols); digits=5)
```

```julia
round.(vcov(ols); digits=5)
```

```julia
pred = DataFrame( x̂ = 1:1:3 );
pred.ŷ = round.(predict(ols), digits=5);

using LaTeXStrings

in_real_numbers = L"\in \mathcal{R}";

plot(xlabel="x", ylabel="y, ŷ $in_real_numbers", legend=:bottomright)
plot!(data.X, data.Y, label="(X,Y) data", seriestype=:scatter)
plot!(pred.x̂, pred.ŷ, label="(x̂,ŷ) model", linewidth=3)

```
