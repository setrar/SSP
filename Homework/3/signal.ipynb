```julia
fs = 1000  # Sampling frequency in Hz
t = 0:1/fs:1  # Time vector

f = 5  # Frequency of the sine wave in Hz
x = sin.(2π * f * collect(t))  # Convert t to an array

plot(t, x, xlabel="Time (s)", ylabel="Amplitude", label="Sine Wave")
```

```julia
using DSP
using Plots

# Generate a signal
fs = 1000  # Sampling frequency in Hz
t = 0:1/fs:1  # Time vector
f_signal = 50  # Frequency of the signal in Hz
signal = sin.(2π * f_signal * t)

# Design a lowpass filter
cutoff_frequency = 100  # Cutoff frequency of the filter in Hz
filter_order = 64  # Order of the FIR filter

filter_coefficients = fir_lowpass(cutoff_frequency, fs, filter_order)

# Apply the lowpass filter to the signal
filtered_signal = conv(signal, filter_coefficients)

# Plot original and filtered signals
plot(t, signal, label="Original Signal", xlabel="Time (s)", ylabel="Amplitude", linewidth=2)
plot!(t, filtered_signal[1:length(t)], label="Filtered Signal", linewidth=2)
```
