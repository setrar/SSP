```julia
using Plots, LaTeXStrings; gr()
```

```julia
f₂(x) = (x<0 ? x+1 : 1-x)*(abs(x)<1 ? 1 : 0)
a,b=-1.5,1.5
Δ = 0.01


F(x) = sum([f₂(u)*Δ for u in a:Δ:x])
xGrid = a:Δ:b
y = [F(u) for u in xGrid]
plot(xGrid, y, c=:blue, xlims=(a,b), ylims=(0,1),
xlabel=L"x", ylabel=L"F(x)", legend=:none)
```

```julia
xGrid = 0:0.01:10
uGrid = 0:0.01:1
busy = 0.8

F(t)= t<=0 ? 0 : 1 - busy*exp(-(1-busy)t)
infimum(B) = isempty(B) ? Inf : minimum(B)
F⁻¹(u) = infimum(filter((x) -> (F(x) >= u),xGrid))

p1 = plot(xGrid,F.(xGrid), c=:blue, xlims=(-0.1,10), ylims=(0,1),
xlabel=L"x", ylabel=L"F(x)")

p2 = plot(uGrid,F⁻¹.(uGrid), c=:blue, xlims=(0,0.95), ylims=(0,maximum(xGrid)),
xlabel=L"u", ylabel=L"F^{-1}(u)")

plot(p1, p2, legend=:none, size=(800, 400))
```
