```julia
using Plots

# Define the parameter space
theta = range(-5, stop=5, length=1000)

# Known standard deviation
sigma = 1.0

# Cramr-Rao bound for the mean of a normal distribution
cramer_rao_bound = fill(1 / (length(theta) * sigma^2), length(theta))

# Plotting the Cramr-Rao bound
plot(theta, cramer_rao_bound, label="Cramr-Rao Bound", color=:red,
     xlabel="Mean ()", ylabel="Variance of Estimator",
     title="Cramr-Rao Bound for the Mean of a Normal Distribution",
     legend=:topleft)
```

The orthogonality principle in statistics refers to the relationship between the residuals (the differences between observed and predicted values) and the predictor variables in linear regression. The principle states that the residuals are orthogonal (uncorrelated) to the predictor variables. In other words, the sum of the products of the residuals and the predictor variables is zero.

Let's consider a simple linear regression example and plot the orthogonality principle. In this case, we'll use the `Plots` library in Julia:
