```julia
N = 256 # Replace with the desired window length
𝑓₁, 𝑓₂ = 0.057, 0.082
𝐴₁, 𝐴₂ = 20, 20
ϕ₁, ϕ₂ = -2π * randn(), 2π * randn()
vₖ = randn();

k = 0:(N-1)
𝑓ₛ = 1024

# Signal
yₖ = 𝐴₁ .* cos.(2π * 𝑓₁ .* k ./ 𝑓ₛ .+ ϕ₁) .+ vₖ

# FFT
𝑌ₖ = 𝓕(yₖ)
frequencies = fftfreq(N, 1 / 𝑓ₛ)  # calculate frequency bins for the FFT output

# Plotting
plot(frequencies[1:Int(N/2)], 20 * log10.(abs.(𝑌ₖ[1:Int(N/2)]));
    label="Magnitude Spectrum"
    , xlabel="Frequency (Hz)", ylabel="Magnitude"
    , title="Magnitude Spectrum of y_k"
)
```
