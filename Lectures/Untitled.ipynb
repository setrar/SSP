```julia
using DSP
using Plots

# Function to create a Gaussian pulse
function gaussian_pulse(t, t0, f0)
    return exp.(-Ï€ * ((t .- t0) * f0).^2)
end

# Time vector
fs = 1000  # Sampling frequency
t = 0:1/fs:1  # Time vector

# Create a Gaussian pulse
t0 = 0.5  # Center of the pulse in time
f0 = 10   # Center frequency of the pulse
signal = gaussian_pulse(t, t0, f0)

# Compute the Short-Time Fourier Transform (STFT)
f, t, Zxx = stft(signal, 256, 200)  # 256-point FFT with 200-point overlap

# Plot the real and imaginary parts separately
plot(t, real.(Zxx), xlabel="Time (s)", ylabel="Real Part", label="STFT Real Part")
# plot!(t, imag.(Zxx), xlabel="Time (s)", ylabel="Imaginary Part", label="STFT Imaginary Part")
```

```julia
using DSP
using Plots

# Number of subcarriers
N = 64

# Generate random OFDM symbol
ofdm_symbol = randn(N)

# Create a polyphase filter bank
filters = randn(4, N)  # Example: 4 filters, each of length N

# Apply the polyphase filter bank
filtered_symbols = filters * ofdm_symbol

# Plot original and filtered symbols
plot(ofdm_symbol, xlabel="Subcarrier Index", ylabel="Amplitude", label="Original OFDM Symbol")
plot!(filtered_symbols, label="Filtered OFDM Symbols", linestyle=:dash)
```
