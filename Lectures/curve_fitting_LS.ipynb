```julia
using LsqFit
```

```julia
# Define the model function
model(x, θ) = θ[1] * sin.(θ[2] .* x)
```

```julia
# Generate some example data
x = 0:0.1:4π
Y = 2.0 * sin.(0.5 .* x) + 0.1 * randn(size(x))

# Initial guess for parameters
θ = [1.0, 0.6]
```

```julia
# Fit the model to the data using LsqFit
fit_result = curve_fit(model, x, Y, θ)

# Extract the fitted parameters
θ̂ = coef(fit_result)

# Generate fitted curve using the optimized parameters
Ŷ = model(x, θ̂);
```

```julia
# Display the optimized parameters
println("Fitted Parameters: ", θ̂)
```

```julia
using Plots
```

```julia
# Plot the data and the fitted curve
plot(x, Y, seriestype=:scatter, label="Data")
plot!(x, Ŷ, label="Fitted Curve")
```
