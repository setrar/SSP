```julia
using StatsPlots: plot
using Distributions: Normal, pdf
```

```julia
ùëµ(m,sd) = Normal(m, sd)
```

```julia
# Define multiple probability distributions
N‚ÇÅ = ùëµ(0, 1)
N‚ÇÇ = ùëµ(2, 0.5)
N‚ÇÉ = ùëµ(-2, 0.8);
```

```julia
# Generate sample data from the distributions
X = range(-5.0, stop=5, length=100)
y‚ÇÅ = pdf.(N‚ÇÅ, X)
y‚ÇÇ = pdf.(N‚ÇÇ, X)
y‚ÇÉ = pdf.(N‚ÇÉ, X);
```

```julia
# Plot the densities
plot( X, y‚ÇÅ, label="Distribution N‚ÇÅ", linewidth=2)
plot!(X, y‚ÇÇ, label="Distribution N‚ÇÇ", linewidth=2)
plot!(X, y‚ÇÉ, label="Distribution N‚ÇÉ", linewidth=2, legend =:topright)

# Customize the plot
xlabel!("X-axis")
ylabel!("Probability Density")
title!("Multiple Probability Distributions")
```
