```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

from plot_rea import *
from pylab import *
```

```python
M = 20   # number of bins in histograms
N = 2500 # Number of samples per realization 
K = 200  # Total number of realizations

XGauss = stats.norm(loc=0, scale=1)

something = np.outer( 
      np.ones((K, 1))
    , np.sin(2 * np.pi * np.arange(N) / N)
)

#Sine wave plus noise
X = 3 * XGauss.rvs(size=(K, N)) + 3 * something

print("Standard deviation of time averages: ", np.std(np.mean(X, axis=1))) 

# pylab.rcParams['figure.figsize'] = (10.0, 8.0)

plt.rcParams['figure.figsize'] = (8, 5) 

plot_rea (X, nb=M, fig =1)
```

```python
def square(x):
    """square(x): \n
    Returns a pulse train with period :math:`2\pi`
    """
    return sign(sin(x))
```

```python
K = 1000
N = 1000
fo = 2.2 / N
S = zeros((K, N))
for r in range(K):
    S[r, :] = 1.1 * square(2 * pi * fo * arange(N) + 2 * pi * rand(1, 1))
plot_rea(S, fig=2)
```
