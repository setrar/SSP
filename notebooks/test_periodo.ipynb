```julia
N = 100
x = range(-2π, stop=2π, length=N) # In Matlab x = linspace(-2*pi, 2*pi);
pl1 = plot(x, sin.(x) .+ 1/2 .* rand(length(x)))

t, y = periodo(x,Int(N/10))
pl2 = plot(t,10*log10.(abs2.(y));
    xlabel = "Normalized frequency", ylabel = "dB"
    , grid = true
)
plot(pl1,pl2;
    layout = (2, 1)
    , size = ( 600, 800)
    , margin = 5mm
)
```
